```python
import pandas as pd
import pandas_datareader as pdr
import numpy as np
import openpyxl
import yfinance as yf
import quandl
import datetime
import natsort

from openpyxl.workbook import Workbook
from openpyxl import load_workbook
from pandas_datareader import data
from datetime import datetime, timezone
from natsort import index_natsorted

pd.core.common.is_list_like = pd.api.types.is_list_like

# Tickers List below;
BTCUSD = "BTC-USD"
GBPUSD = "GBPUSD=X"
USDJPY = "JPY=X"
XAUUSD = "XAUUSD=X"
SOLUSD = "SOL-USD"
UK100 = "^FTSE"
DJ30 = "^DJI"
SPX500 = "^SPX"
NSDQ100 = "NQ=F"

Ticker = BTCUSD


# Create MONTHLY dataframe & excel sheet
data_Month = yf.download(Ticker, period = "max", interval = "1mo", group_by = "ticker", auto_adjust = True)

data_Month.index = data_Month.index.tz_localize(None)

data_Month.pop("Volume")

# data_Month["Open"] = data_Month["Open"]/100
# data_Month["High"] = data_Month["High"]/100
# data_Month["Low"] = data_Month["Low"]/100
# data_Month["Close"] = data_Month["Close"]/100

data_Month_sorted = data_Month.sort_values(by='Date', ascending=False)

data_Month_sorted.to_excel("BTC_Month.xlsx")


# Create WEEKLY dataframe & excel sheet
data_Week = yf.download(Ticker, period = "max", interval = "1wk", group_by = "ticker", auto_adjust = True)

data_Week.index = data_Week.index.tz_localize(None)

data_Week.pop("Volume")

data_Week_sorted = data_Week.sort_values(by='Date', ascending=False)

data_Week_sorted.to_excel("BTC_Week.xlsx")


# Create DAILY dataframe & excel sheet
data_Daily = yf.download(Ticker, period = "max", interval = "1d", group_by = "ticker", auto_adjust = True)

data_Daily.index = data_Daily.index.tz_localize(None)

data_Daily.pop("Volume")

data_Daily_sorted = data_Daily.sort_values(by='Date', ascending=False)

data_Daily_sorted.to_excel("BTC_Daily.xlsx")


# # Create 4-HOURLY dataframe & excel sheet
# # 4 Hour charts do not get downloaded with the yfinance python downloader
# data_4_Hourly = yf.download(Ticker, period = "ytd", interval = "4h", group_by = "ticker", auto_adjust = True)

# data_4_Hourly.index = data_4_Hourly.index.tz_localize(None)

# data_4_Hourly.pop("Volume")

# # data_Hourly.at[1, 1] = "Date"
# # data_Hourly.columns = ['Date', 'Open', 'High', 'Low', 'Close']
# # data_Hourly.rename(columns = {"" : "Date"}, inplace = True)
# # data_Hourly.columns.values[0] = 'Date'
# # data_Hourly.set_index('Date')
# # data_Hourly = data_Hourly.rename(columns={'Unnamed: 0': 'Date'})

# # data_Hourly_sorted = data_Hourly.sort_values(by='Date', ascending=False)

# ## The Hourly Dataframe is sorted with a different code because it is missing the "Date" label on the first column, so first colum is now an index column.
# data_4_Hourly_sorted = data_4_Hourly.sort_index(ascending = False)

# data_4_Hourly_sorted.to_excel("BTC_4Hr.xlsx")


# Create HOURLY dataframe & excel sheet
# the 1 Hour chart can only be downloaded up to a period of 730 days, so we use a period of 23 months
data_Hourly = yf.download(Ticker, period = "730d", interval = "1h", group_by = "ticker", auto_adjust = True)

data_Hourly.index = data_Hourly.index.tz_localize(None)

data_Hourly.pop("Volume")

# data_Hourly.at[1, 1] = "Date"
# data_Hourly.columns = ['Date', 'Open', 'High', 'Low', 'Close']
# data_Hourly.rename(columns = {"" : "Date"}, inplace = True)
# data_Hourly.columns.values[0] = 'Date'
# data_Hourly.set_index('Date')
# data_Hourly = data_Hourly.rename(columns={'Unnamed: 0': 'Date'})

# data_Hourly_sorted = data_Hourly.sort_values(by='Date', ascending=False)

## The Hourly Dataframe is sorted with a different code because it is missing the "Date" label on the first column, so first colum is now an index column.
data_Hourly_sorted = data_Hourly.sort_index(ascending = False)

data_Hourly_sorted.to_excel("BTC_Hourly.xlsx")


# Create 30min dataframe & excel sheet
data_30min = yf.download(Ticker, period = "60d", interval = "30m", group_by = "ticker", auto_adjust = True)

data_30min.index = data_30min.index.tz_localize(None)

data_30min.pop("Volume")

data_30min_sorted = data_30min.sort_values(by='Datetime', ascending=False)

data_30min_sorted.to_excel("BTC_30min.xlsx")


# Create 15min dataframe & excel sheet
data_15min = yf.download(Ticker, period = "60d", interval = "15m", group_by = "ticker", auto_adjust = True)

data_15min.index = data_15min.index.tz_localize(None)

data_15min.pop("Volume")

data_15min_sorted = data_15min.sort_values(by='Datetime', ascending=False)

data_15min_sorted.to_excel("BTC_15min.xlsx")


# Create 5min dataframe & excel sheet
data_5min = yf.download(Ticker, period = "60d", interval = "5m", group_by = "ticker", auto_adjust = True)

data_5min.index = data_5min.index.tz_localize(None)

data_5min.pop("Volume")

data_5min_sorted = data_5min.sort_values(by='Datetime', ascending=False)

data_5min_sorted.to_excel("BTC_5min.xlsx")


# Create 1min dataframe & excel sheet
# the 1min chart has a maximum display period of 7 days, rather than 1 month
data_1min = yf.download(Ticker, period = "7d", interval = "1m", group_by = "ticker", auto_adjust = True)

data_1min.index = data_1min.index.tz_localize(None)

data_1min.pop("Volume")

data_1min_sorted = data_1min.sort_values(by='Datetime', ascending=False)

data_1min_sorted.to_excel("BTC_1min.xlsx")



# display data
print("\nDisplay dataframe with data - 1 MONTH")
print(data_Month_sorted)

print("\nDisplay dataframe with data - 1 WEEK")
print(data_Week_sorted)

print("\nDisplay dataframe with data - 1 DAY")
print(data_Daily_sorted)

# print("\nDisplay dataframe with data - 4 HOUR")
# print(data_4_Hourly_sorted)

print("\nDisplay dataframe with data - 1 HOUR")
print(data_Hourly_sorted)

print("\nDisplay dataframe with data - 30 MINUTE")
print(data_30min_sorted)

print("\nDisplay dataframe with data - 15 MINUTE")
print(data_15min_sorted)

print("\nDisplay dataframe with data - 5 MINUTE")
print(data_5min_sorted)

print("\nDisplay dataframe with data - 1 MINUTE")
print(data_1min_sorted)
```
